```python
import torch
import torchvision
import numpy as np
from torch.utils.data import TensorDataset, DataLoader
transform = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize((0.1307,), (0.3081,))
])
```

```python
train_dataset = torchvision.datasets.MNIST(root='./', train=True, transform=transform, download=True)
test_dataset = torchvision.datasets.MNIST(root='./', train=False, transform=transform, download=True)

included_classes = [1, 2]
train_filter = np.isin(train_dataset.targets, included_classes)
test_filter = np.isin(test_dataset.targets, included_classes)

train_dataset.targets = train_dataset.targets[train_filter]
train_dataset.data = train_dataset.data[train_filter]
test_dataset.targets = test_dataset.targets[test_filter]
test_dataset.data = test_dataset.data[test_filter]
```

```python
ones = X_train[(y_train == 1)]
```

```python
p_b_ones = ones.sum(axis=0)/ones.sum()
```

```python
p_ones = (y_train==1).sum()/len(y_train)
p_twos = (y_train == 2).sum() / len(y_train)
```

```python
def pred(p_ones,p_b,p_b_ones,p_twos,y_pred):
  return ((p_b_ones*y_pred)[y_pred>0].prod()*p_ones)/(y_pred*p_b)[y_pred!=0].prod()
```

```python
y_pred = []
for i in X_test.reshape(-1,28*28):
  y_pred.append(pred(p_ones,p_b,p_b_ones,p_twos,i).item())
```

```python
a = []
for i in y_pred:
  if i >0:
    a.append(1)
  else:
    a.append(2)
```

```python
from sklearn.metrics import accuracy_score as auc
```

```python
auc(y_test,a)
```
