```python

import os
import time
import glob

from tqdm.notebook import tqdm
from collections import defaultdict

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

import torch
from torch import nn
import torch.nn.functional as F
from torchsummary import summary

import torchvision
from torchvision import transforms
import torch.nn.functional as F
from IPython.display import clear_output
%matplotlib inline
sns.set_style(style='white')
```

```python
from torchvision.utils import make_grid
stats = (0.5, 0.5, 0.5), (0.5, 0.5, 0.5)
def show_images(images, nmax=64):
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_xticks([]); ax.set_yticks([])
    ax.imshow(make_grid(denorm(images.detach()[:nmax]), nrow=8).permute(1, 2, 0))

def show_batch(dl, nmax=64):
    for images, _ in dl:
        show_images(images, nmax)
        break
def denorm(img_tensors):
    return img_tensors * stats[1][0] + stats[0][0]
```

```python
data_dir = "data"
data_image_paths = glob.glob(f"{data_dir}/*/*.png")
data_image_labels = [path.split('/')[-2] for path in data_image_paths]
train_files_path, val_files_path = train_test_split(
    data_image_paths,
    test_size=0.1,
    stratify=data_image_labels
)
print(
    f"Изображений в train: {len(train_files_path)}\nИзображений в val: {len(val_files_path)}"
)
```

```python
from torchvision.datasets.folder import default_loader
def is_valid_file(filepath):
    if not filepath.endswith('.jpg') and not filepath.endswith('.jpeg') and not filepath.endswith('.png'):
        return False
    try:
        default_loader(filepath)
        return True
    except:
        return False

input_size =64

train_transform = transforms.Compose([
    transforms.Resize(input_size),
    transforms.CenterCrop(input_size),
    #transforms.ColorJitter(0.9, 0.9, 0.9),
    #transforms.RandomAffine(5),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

train_dataset = torchvision.datasets.ImageFolder(
    root=data_dir,
    transform=train_transform,
    is_valid_file = lambda x: x in train_files_path
)
```

```python
Discriminator_Anime = nn.Sequential(

    nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(64),
    nn.LeakyReLU(0.2, inplace=True),


    nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.LeakyReLU(0.2, inplace=True),


    nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.LeakyReLU(0.2, inplace=True),


    nn.Conv2d(256, 512, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(512),
    nn.LeakyReLU(0.2, inplace=True),


    nn.Conv2d(512, 1, kernel_size=4, stride=1, padding=0, bias=False),

    nn.Flatten(),
    nn.Sigmoid())
```

```python
alpha = 128
Generator_Anime = nn.Sequential(
    nn.ConvTranspose2d(alpha, 512, kernel_size=4, stride=1, padding=0, bias=False),
    nn.BatchNorm2d(512),
    nn.ReLU(True),

    nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.ReLU(True),

    nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.ReLU(True),

    nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(64),
    nn.ReLU(True),

    nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1, bias=False),
    nn.Tanh()

)
```

```python
from torchsummary import summary
#model = Generator_Anime()
summary(Discriminator_Anime, input_size=(3,64, 64), device='cpu')
```

```python
# loss
import torch, gc
gc.collect()
torch.cuda.empty_cache()
dev = 'cuda'
G = Generator_Anime.to(dev)
D = Discriminator_Anime.to(dev)
# optimizer
lr = 0.0002
G_optimizer = torch.optim.Adam(G.parameters(), lr = lr )
D_optimizer = torch.optim.Adam(D.parameters(), lr = lr )
```

```python
def D_train(x):
    #=======================Train the discriminator=======================#
    D.zero_grad()

    # train discriminator on real
    x_real, y_real = x, torch.ones(BATCH_SIZE, 1)
    x_real, y_real = torch.autograd.Variable(x_real.to(dev)), torch.autograd.Variable(y_real.to(dev))

    D_output = D(x_real)
    D_real_loss = F.binary_cross_entropy(D_output, y_real)
    D_real_score = D_output

    # train discriminator on facke
    z = torch.autograd.Variable(torch.randn(BATCH_SIZE,alpha,1,1).to(dev))
    x_fake, y_fake = G(z).reshape(BATCH_SIZE,3, 64,64) ,torch.autograd.Variable(torch.zeros(BATCH_SIZE, 1).to(dev))

    D_output = D(x_fake)
    D_fake_loss = F.binary_cross_entropy(D_output, y_fake)
    D_fake_score = D_output

    # gradient backprop & optimize ONLY D's parameters
    D_loss = D_real_loss + D_fake_loss
    D_loss.backward()
    D_optimizer.step()

    return  D_loss.data.item()
```

```python
def G_train(x):
    #=======================Train the generator=======================#
    G.zero_grad()

    z = torch.autograd.Variable(torch.randn(BATCH_SIZE,alpha, 1,1).to(dev))

    y = torch.autograd.Variable(torch.ones(BATCH_SIZE, 1).to(dev))

    G_output = G(z).reshape(BATCH_SIZE,3, 64,64)
    D_output = D(G_output)
    G_loss = F.binary_cross_entropy(D_output, y)

    # gradient backprop & optimize ONLY G's parameters
    G_loss.backward()
    G_optimizer.step()

    return G_loss.data.item()
```

```python
BATCH_SIZE = 128
dl= torch.utils.data.DataLoader(
    train_dataset, batch_size=BATCH_SIZE, shuffle=True
)
```

```python
n_epoch = 100
for epoch in range(1, n_epoch+1):
    D_losses, G_losses = [], []
    for batch_idx, (x, _) in enumerate(dl):
        if len(x) != BATCH_SIZE:
            continue
        D_losses.append(D_train(x))
        G_losses.append(G_train(x))

    print('[%d/%d]: loss_d: %.3f, loss_g: %.3f' % (
            (epoch), n_epoch, torch.mean(torch.FloatTensor(D_losses)), torch.mean(torch.FloatTensor(G_losses))))
```
